```python
import pandas as pd
import numpy as np
```

## Read in data 

```python
contents = []
```

```python
#parse line by line, and strip contents in line 
with open('pre-discussion-4125.txt', encoding='utf8') as f:
    for line in f:
        line = line.strip() #strip line
        stuff = [x.strip() for x in line.split("\t")] #extract data, split by tab
        contents.append(stuff)
    f.close() #close reading 
```

```python
contents
```

```python
np.asarray(contents) #convert list to numpy array
```

```python
#convert numpy array to pandas dataframe 
xiahua = pd.DataFrame(data=contents, columns = ['Project ID'
                                                , 'Discussion ID'
                                               , 'Date'
                                               , 'Reply'
                                               , 'Discussant ID'
                                               , 'Discussion Text'
                                               , 'Days Before Crowdfunding'])
```

```python
xiahua.drop(xiahua.index[0], inplace=True)
```
